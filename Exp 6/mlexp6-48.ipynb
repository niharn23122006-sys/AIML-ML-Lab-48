```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score,accuracy_score
```

```python
data = pd.read_csv("credit.csv" , skiprows=1)
data = data.sample(5000 , random_state = 42 )
print(data.columns)
```

```python
y = data ['default payment next month']
x = data [['BILL_AMT1','BILL_AMT2']]
print(x.dtypes)
print(x.head())
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size = 0.3, random_state=42 )
scaler= StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
models = {
    "Linear SVM ": SVC(kernel='linear', C= 1, class_weight='balanced'),
    "Polynomial SVM" : SVC (kernel='poly',
                           degree=2,
                            C=1,gamma='scale',
                            class_weight='balanced'
                        ),
    "RBF SVM" : SVC(kernel = 'rbf',
                    C=5,
                    gamma = 0.1,
                    class_weight = 'balanced'
                   )
}

for name, model in models.items():
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    y_pred_train = model.predict(x_train)

    print("\n",name)
    print("Confusion Matrix",confusion_matrix(y_test,y_pred))
    print("precision score",precision_score(y_test,y_pred, zero_division=0))
    print("recall score",recall_score(y_test,y_pred))
    print("f1 score",f1_score(y_test,y_pred))
    print("Accuracy score :", accuracy_score(y_test , y_pred))
    print("Accuracy score :", accuracy_score(y_train , y_pred_train))
```

```python
def plot_boundary(model, title, X, y):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, h),
        np.arange(y_min, y_max, h)
    )
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    # Class 0
    plt.scatter(
        X[y == 0, 0],
        X[y == 0, 1],
        label='No Default (0)',
        marker='o',
        edgecolors='k'
    )
    # Class 1
    plt.scatter(
        X[y == 1, 0],
        X[y == 1, 1],
        label='Default (1)',
        marker='x',
        edgecolors='k'
    )
    plt.legend()
    plt.title(title)
    plt.xlabel("BILL_AMT1 (scaled)")
    plt.ylabel("BILL_AMT2 (scaled)")
```

```python
plt.figure(figsize=(12, 4))

for i, (name, model) in enumerate(models.items()):
    plt.subplot(1, 3, i + 1)
    model.fit(x_train, y_train)
    plot_boundary(model, name, x_train, y_train)

plt.tight_layout()
plt.show()
```
