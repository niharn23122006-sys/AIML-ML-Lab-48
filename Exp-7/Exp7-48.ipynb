## Exp - 7: Ensemble Learning for House Price Prediction

```python
import time 
import pandas as pd
from sklearn.datasets import fetch_california_housing 
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor 
from sklearn.ensemble import GradientBoostingRegressor 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
data = fetch_california_housing()
print(data.keys())
```

```python
x = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

print(x.head())
print("Samples:", len(x))
print(y[:23])
```

## Decision Tree

```python
x_train, x_test, y_train, y_test = train_test_split(
    x , y ,test_size = 0.2 , random_state=42
)

tree = DecisionTreeRegressor()

start = time.perf_counter()
tree.fit(x_train, y_train)
tree_time = time.perf_counter() - start 

pred_tree_test = tree.predict(x_test)
err_tree_test = mean_absolute_error(y_test, pred_tree_test)
pred_tree_train = tree.predict(x_train)
err_tree_train = mean_absolute_error(y_train, pred_tree_train)

print("Decision Tree Train MAE:" , err_tree_train)
print("Decision Tree Test MAE:" , err_tree_test)
print("Decision Tree Time:", tree_time)
```

## Random Forest

```python
x_train, x_test, y_train, y_test = train_test_split(
    x , y ,test_size = 0.2 , random_state=42
)

rf = RandomForestRegressor()

start = time.perf_counter()
rf.fit(x_train, y_train)
rf_time = time.perf_counter() - start 

pred_rf_test = rf.predict(x_test)
err_rf_test = mean_absolute_error(y_test, pred_rf_test)
pred_rf_train = rf.predict(x_train)
err_rf_train = mean_absolute_error(y_train, pred_rf_train)

print("Random forest-bagging Train MAE:" , err_rf_train)
print("Random forest-bagging  Test MAE:" , err_rf_test)
print("Random forest-bagging  Time:", rf_time)
```

## Gradient Boosting

```python
x_train, x_test, y_train, y_test = train_test_split(
    x , y ,test_size = 0.2 , random_state=42
)

gb = GradientBoostingRegressor(n_estimators=300, learning_rate=0.1, max_depth=5)

start = time.perf_counter()
gb.fit(x_train, y_train)
gb_time = time.perf_counter() - start 

pred_gb_test = gb.predict(x_test)
err_gb_test = mean_absolute_error(y_test, pred_gb_test)
pred_gb_train = gb.predict(x_train)
err_gb_train = mean_absolute_error(y_train, pred_gb_train)

print("Gradient Boosting (Boosting) Train MAE:" , err_gb_train)
print("Gradient Boosting (Boosting)  Test MAE:" , err_gb_test)
print("Gradient Boosting (Boosting)  Time:", gb_time)
```

```python
print("\n ---- Comparison For the Test MAE ----")
print("Decision Tree Test:", err_tree_test)
print("Random Forest test :", err_rf_test)
print("Gradient Boosting Test:", err_gb_test)
```

```python
print("\n ---- Comparison For the Train MAE ----")
print("Decision Tree Test:", err_tree_train)
print("Random Forest test :", err_rf_train)
print("Gradient Boosting Test:", err_gb_train)
```

```python
print("\n ---- Comparison For the Total Time ----")
print("Decision Tree Test:", tree_time)
print("Random Forest test :", rf_time)
print("Gradient Boosting Test:", gb_time)
```
